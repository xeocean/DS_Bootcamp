## Упражнение 04: Обогащение и преобразования

```python
import pandas as pd
import numpy as np
```

## read the JSON file 

```python
pd.options.display.float_format = '{:.2f}'.format
df = pd.read_json('../data/auto.json')
df
```

## enrich the dataframe using a sample from that dataframe

```python
sample_df = df.sample(n=200, random_state=21)
sample_df
```

```python
con_cat_rows = pd.concat([df, sample_df], ignore_index=True)

con_cat_rows
```

## enrich the dataframe with the data from another dataframe

```python
np.random.seed(21)

con_cat_rows['Year'] = pd.Series([np.random.randint(1980, 2019) for i in range(len(con_cat_rows))], name='Year')

con_cat_rows
```

## enrich the dataframe concat_rows by a new column with the data generated

```python
surname_df = pd.read_json('../data/surname.json')
surname_df.columns = surname_df.iloc[0]
surname_df.drop(0, inplace=True)

surname_df
```

```python
car_numbers = con_cat_rows.drop_duplicates('CarNumber')
car_numbers
```

```python
surname_df = surname_df.sample(n=len(car_numbers), replace=True, random_state=21)

owners = pd.DataFrame({'CarNumber': car_numbers['CarNumber'].values, 'SURNAME': surname_df['NAME'].values})

owners
```

```python
data = [ 
    ["70RUS", 1.0, 100.0, "Dodge", "Neon", 2001],
    ["71RUS", 2.0, 1100.0, "Ford", "Focus", 2000],
    ["72RUS", 1.0, 200.0, "Ford", "Focus", 1999],
    ["73RUS", 5.0, 500.0, "Ford", "Focus", 2004],
    ["74RUS", 6.0, 500.0, "Ford", "Focus", 2004],
    ]
observations_df = pd.DataFrame(data, columns=con_cat_rows.columns)

observations_df
```

```python
fines = pd.concat([con_cat_rows, observations_df])
fines
```

```python
owners.drop(owners.tail(20).index, inplace=True)
owners
```

```python
data = [ 
    ["80RUS", "Neon"],
    ["81RUS", "Neo"],
    ["82RUS", "Ne"],
    ]

observations_more_df = pd.DataFrame(data, columns=owners.columns)
observations_more_df.set_index('SURNAME')

owners = pd.concat([observations_more_df, owners])
owners
```

```python
inner_df = pd.merge(left=fines, right=owners, on='CarNumber', how='inner')
inner_df
```

```python
other_df = pd.merge(left=fines, right=owners, on='CarNumber', how='outer')
other_df
```

```python
left_df = pd.merge(left=fines, right=owners, on='CarNumber', how='left')
left_df
```

```python
right_df = pd.merge(left=fines, right=owners, on='CarNumber', how='right')
right_df
```

## create a pivot table from the fines dataframe, it should look like this (the values are the sums of the fines), but with all the years 

```python
table_df = pd.pivot_table(data=fines, columns='Year', values='Fines', index=['Make', 'Model'], aggfunc='sum')
table_df
```

## save both the fines and owners dataframes to CSV files without an index

```python
fines.to_csv('../data/fines.csv', index=False)
owners.to_csv('../data/owners.csv', index=False)
```
