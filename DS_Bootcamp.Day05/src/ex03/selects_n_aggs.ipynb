## Упражнение 03: Выборки и агрегации

```python
import pandas as pd
```

## load the JSON file

```python
df = pd.read_json('../data/auto.json', orient='records')
df.set_index('CarNumber', inplace=True)
df
```

## make the following selects

```python
df[df['Fines'] > 2100]
```

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)] 
```

```python
search_list = ['Focus', 'Corolla']
df[df['Model'].isin(search_list)]
```

```python
search_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(search_list)]
```

## make the aggregations with the make and the model

```python
df.groupby('Make')['Fines'].median()
```

```python
df.groupby(['Make', 'Model'])['Fines'].median()
```

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

```python
df.groupby(['Make', 'Model'])['Fines'].std()
```

## make the aggregations with the car number

```python
df.groupby('CarNumber')['Fines'].count().sort_values(ascending=False)
```

```python
df[df.index == 'Y7689C197RUS']
```

```python
df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
```

```python
df[df.index == 'X758HY197RUS']
```

```python
car_numbers = df.groupby('CarNumber')['Model'].nunique()
car_multiple = car_numbers[car_numbers > 1]

df[df.index.isin(car_multiple.index)]
```
