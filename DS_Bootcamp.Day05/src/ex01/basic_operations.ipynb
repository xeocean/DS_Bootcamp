## Упражнение 01: Основные операции

```python
import pandas as pd
```

## create a dataframe views with two columns: datetime and user

```python
df = pd.read_csv('../data/feed-views.log', names=['datetime', 'user'], parse_dates=['datetime'], sep='\t')

df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second

df.info()
df
```

## create the new column daytime

```python
df.set_index('user', inplace=True)
df['daytime'] = pd.cut(df['hour'], bins=[0, 4, 7, 11, 17, 20, 24], labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening'], right=False)

df
```

## calculate the number of elements in your dataframe

```python
total = df.shape[0]
print(f'Total: {total}')
df['daytime'].value_counts()
```

## sort values in your dataframe by hour, minute, and second in ascending order

```python
df.sort_values(['hour', 'minute', 'second'])
```

## calculate the minimum and maximum for the hours and the mode for the daytime categories

```python
night_df = df[df['daytime'] == 'night']
morning_df = df[df['daytime'] == 'morning']

max_night = night_df['hour'].max()
min_morning = morning_df['hour'].min()

print(f'Max hour night: {max_night}')
print(f'Min hour morning: {min_morning}')


users_night = df[df['hour'] == max_night]
users_morning = df[df['hour'] == min_morning]

print(users_night)
print(users_morning)

print(f'Moda for hour: {df['hour'].mode()}')
print(f'Moda for daytime: {df['daytime'].mode()}')
```

## show the 3 earliest hours in the morning and the corresponding usernames and the 3 latest hours and the usernames using nsmallest() and nlargest()

```python
morning_df = df[df['daytime'] == 'morning']
print(morning_df.nsmallest(3, columns=['hour']))
print(morning_df.nlargest(3, columns=['hour']))
```

## use the method describe() to get the basic statistics for the columns

```python
iqr = df.describe()

print(iqr)

q1 = iqr['hour']['25%']
q3 = iqr['hour']['75%']

print(f'Interquartile range: {q3 - q1}')
```
