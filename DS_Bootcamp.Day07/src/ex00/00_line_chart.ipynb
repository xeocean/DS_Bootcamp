## Exercise 00 : Line chart

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
query = '''
        select uid, datetime from pageviews where uid like 'user_%';
        '''
df = pd.read_sql(query, conn, parse_dates='datetime')
```

```python
df_view = df.groupby(df['datetime'].dt.date).count()
df_view = df_view['datetime'].rename('view')
df_view.index.name = 'date'
df_view
```

```python
df_view.plot(x='date', y='view', figsize=(15, 8), rot=90, fontsize=8, title='Views per day')
```

```python
conn.close()
```
